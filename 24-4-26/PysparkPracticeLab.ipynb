```python
products = products.withColumnRenamed("price","product_price")
customers = customers.withColumnRenamed("name","customer_name")

customers.select("customer_name","age").show()
products.select("product_name","category").show()

orders.select("order_id","order_date").show()
payments.select("payment_type","amount").show()

from pyspark.sql.functions import col

products = products.withColumn("price_in_lakhs", col("product_price")/100000)

from pyspark.sql.functions import when
customers = customers.withColumn("age_group",
    when(col("age") < 30,"Young")
    .when(col("age") < 40,"Adult")
    .otherwise("Senior"))

order_items.filter(order_items.quantity > 2).show()

customers.orderBy("age").show()
```

```python
customers.orderBy("age").show()
customers.orderBy(col("age").desc()).show()

products.orderBy(col("product_price").desc()).show(5)
products.orderBy("product_price").show(3)

orders.orderBy("order_date").show()
payments.orderBy("amount").show()

payments.orderBy(col("amount").desc()).show(3)
payments.orderBy("amount").show(3)

customers.orderBy("city").show()
products.orderBy("category").show()
```

```python
from pyspark.sql.functions import sum, avg, max, min

payments.select(sum("amount")).show()
payments.select(avg("amount")).show()
payments.select(max("amount")).show()
payments.select(min("amount")).show()

customers.groupBy("city").count().show()
products.groupBy("category").count().show()

products.groupBy("category").avg("product_price").show()

order_items.groupBy("product_id").sum("quantity").show()

customers.select(avg("age")).show()

orders.count()
```

```python
orders.join(customers, "customer_id").show()

orders.join(customers, "customer_id") \
.select("customer_name","status").show()

orders.join(order_items, "order_id").show()

order_items.join(products, "product_id").show()

from pyspark.sql.functions import expr

df = order_items.join(products, "product_id") \
.withColumn("revenue", col("quantity") * col("product_price"))

df.show()

full_df = orders.join(customers,"customer_id") \
.join(order_items,"order_id") \
.join(products,"product_id")

full_df.select("customer_name","product_name","quantity").show()

full_df.groupBy("order_id") \
.sum("quantity","product_price").show()

df.groupBy("product_id").sum("revenue").show()

full_df.groupBy("customer_id").sum("quantity").show()
```

```python
orders.groupBy("customer_id").count().show()

orders.join(customers,"customer_id") \
.groupBy("city").count().show()

df.groupBy("product_name").sum("revenue").show()

df.groupBy("category").sum("revenue").show()

full_df.groupBy("city").sum("product_price").show()

order_items.groupBy("product_id").sum("quantity").show()

df.groupBy("category").sum("quantity").show()

full_df.groupBy("customer_name").sum("product_price").show()

full_df.groupBy("customer_name") \
.sum("product_price").orderBy(col("sum(product_price)").desc()).show(5)

df.groupBy("product_name") \
.sum("revenue").orderBy(col("sum(revenue)").desc()).show(3)
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import rank, dense_rank, row_number

window = Window.orderBy("product_price")

products.withColumn("rank", rank().over(window)).show()

window_cat = Window.partitionBy("category").orderBy("product_price")
products.withColumn("rank", rank().over(window_cat)).show()

window_date = Window.orderBy("signup_date")
customers.withColumn("row_num", row_number().over(window_date)).show()

rev_df = df.groupBy("product_id").sum("revenue")

window_rev = Window.orderBy(col("sum(revenue)").desc())

rev_df.withColumn("rank", rank().over(window_rev)).show()
rev_df.withColumn("dense_rank", dense_rank().over(window_rev)).show()

customers.join(orders,"customer_id") \
.withColumn("rank", rank().over(Window.partitionBy("city").orderBy("customer_id"))).show()

products.withColumn("rank", rank().over(window_cat)).show()

df.withColumn("running_total", sum("revenue").over(Window.orderBy("product_id"))).show()

df.withColumn("running_total", sum("revenue").over(Window.partitionBy("product_id"))).show()

payments.withColumn("rank", rank().over(Window.orderBy("amount"))).show()
```
